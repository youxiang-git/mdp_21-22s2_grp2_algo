```python
import numpy as np
import networkx as nx
import scipy as sc
from matplotlib import pyplot as plt
np.set_printoptions(threshold=np.inf, linewidth=np.inf)
```

```python
grid_graph1 = nx.grid_2d_graph(20,20)
plt.figure(figsize=(20,20))
pos = {(x,y):(y,x) for x,y in grid_graph1.nodes()}
nx.draw(grid_graph1, pos=pos, 
        node_color='lightblue', 
        with_labels=True,
        node_size=600)
```

```python
ignore = [(18, 2)]
pos = {(x,y):(y,x) for (x,y) in grid_graph1.nodes() if (x,y) not in ignore}
grid_graph1.remove_node((18, 2))
plt.figure(figsize=(20,20))
nx.draw(grid_graph1, pos= pos,
    node_color='lightblue', 
    with_labels=True,
    node_size=600)
    

adj_g1 = nx.adj_matrix(grid_graph1)
adj_g1.toarray()
```
